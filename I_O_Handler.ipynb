## I/O Handler

```python
class IOHandler:
    """
    Universal input/output handler for cheminformatics pipelines.
    Handles data loading, output folder management, and logging.
    """

    def __init__(self, input_path=None, output_root=None, project_name=None):
        self.input_path = input_path
        self.project_name = project_name or "Project"
        
        # Default output root: same directory as input or current working dir
        if output_root is None:
            self.output_root = os.path.dirname(input_path) if input_path else os.getcwd()
        else:
            self.output_root = output_root

        # Create a timestamped session folder
        timestamp = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
        self.session_folder = os.path.join(self.output_root, f"{self.project_name}_Results_{timestamp}")
        os.makedirs(self.session_folder, exist_ok=True)

        print(f"📂 Output directory created: {self.session_folder}")

    def load_csv(self, path=None):
        """Load CSV file (default: self.input_path)."""
        target = path or self.input_path
        if not target or not os.path.exists(target):
            raise FileNotFoundError(f"CSV file not found at: {target}")
        df = pd.read_csv(target)
        print(f"✅ Loaded data: {len(df)} rows × {len(df.columns)} columns")
        return df

    def get_subfolder(self, name):
        """Create and return a subfolder inside the main session folder."""
        subfolder = os.path.join(self.session_folder, name)
        os.makedirs(subfolder, exist_ok=True)
        return subfolder

    def save_plot(self, fig, name, subfolder="Plots"):
        """Save matplotlib figure automatically into organized subfolders."""
        folder = self.get_subfolder(subfolder)
        fig_path = os.path.join(folder, f"{name}.png")
        fig.savefig(fig_path, dpi=300, bbox_inches="tight")
        print(f"🖼️ Saved plot: {fig_path}")

    def save_dataframe(self, df, name, subfolder="Data"):
        """Save DataFrame as CSV."""
        folder = self.get_subfolder(subfolder)
        csv_path = os.path.join(folder, f"{name}.csv")
        df.to_csv(csv_path, index=False)
        print(f"📊 Saved data: {csv_path}")
```
